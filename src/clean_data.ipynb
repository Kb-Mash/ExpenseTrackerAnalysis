```python
import pandas as pd
from pathlib import Path

DATA_DIR = Path("../data")

def load_raw_data():
    return pd.read_csv(DATA_DIR / "personal_expense_dataset.csv")
```

```python
df = load_raw_data()

print(df.dtypes)
```

```python
# convert data types
df['Date'] = pd.to_datetime(df['Date'])
print(df['Date'].dtype)
```

```python
print(df.dtypes)
```

```python
print(df.head())
```

```python
df['Amount'] = df['Amount'].astype(float)
```

```python
df['Month'] = df['Date'].dt.month_name()
```

```python
# check missing data
print(df.isnull().sum())
```

```python
# drop missing rows
df = df.dropna()
# remove duplicates
df = df.drop_duplicates()
```
